```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import seaborn as sns
df=pd.read_csv('Iris.csv')
#pd.options.display.max_rows=None
plt.style.use("seaborn-whitegrid")
import warnings
warnings.filterwarnings("ignore")
df
```

```python
x=df[[ 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]#selecting features to use
```

# Finding the optimum number of clusters for k-means classification
ill use elbow method im not sure if there other reliable method ,
and  we locate the number of k in between the sse and numberof k_cluster line
where it curves when we visualize as below:


```python
from sklearn.cluster import KMeans
sse=[]
for i in range(1,13):
    
    km= KMeans(n_clusters=i).fit(x)
  

    sse.append(km.inertia_)
    
plt.plot(range(1, 13), sse)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSe') # Within cluster sum of squares
plt.show()
```

```python
df['Species'].value_counts()
```

```python
# Run K-means clustering on the data
km = KMeans(n_clusters=3).fit(x)

x['cluster'] = km.labels_

# Visualize the clusters using a pairplot
sns.pairplot(data=x, hue='cluster', palette='Set2')
plt.show()
```

```python
# Run K-means clustering on the data
kmeans = KMeans(n_clusters=3).fit(x)

x['cluster'] = kmeans.labels_

# Create a scatter plot of the first two features, colored by cluster
fig, ax = plt.subplots()
scatter = ax.scatter(x.iloc[:,0], x.iloc[:,1], c=x['cluster'], cmap='rainbow')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalwidthCm')
legend1 = ax.legend(*scatter.legend_elements(), loc="upper right", title="Clusters")
ax.add_artist(legend1)

# Add cluster centers to the plot
centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=0.5);

plt.show()
```

```python
cluster_label=np.array(x['cluster'])
cluster_label
```
